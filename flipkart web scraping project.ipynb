```python
#importing essential packages
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
#defines function  to  get title
def get_title(soup):
    try:
        title = soup.find("span",attrs={'class':'mEh187'}).text.strip()
    except AttributeError:
        title="unable  to fetch title"
        return title

#defines function to get price
def get_price(soup):
    try:
        price= soup.find("div",attrs={'class':'Nx9bqj CxhGGd'}).text.strip()
    except AttributeError:
        price = "Unable to fetch price"
        return price

#define function to get ratings
def get_ratings(soup):
    try:
        ratings= soup.find("div",attrs={'class':'XQDdHH _1Quie7'}).text.strip()
    except AttributeError:
        ratings = "Unable to fetch"
        return ratings
```

```python

#get the URL of the site to send request
URL = "https://www.flipkart.com/search?q=wrist+watch+for+men&sid=r18%2Cf13&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_2_12_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_2_12_na_na_na&as-pos=2&as-type=RECENT&suggestionId=wrist+watch+for+men%7CWrist+Watches&requestId=abddfee2-ac4f-4ffb-ac59-9981eb41647a&as-searchtext=wrist%20watch%20for%20men"

#to send HTTP request by using header useragent
HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/127.0.0.0 Safari/537.36','Accept-Language':'en-us,en;q=0.5'})

#send request to  website by using request.get method
webpage = requests.get(URL,headers = HEADERS)

#show if the request  returns response or not if it reponse 200 then it is correct otherwise there is some error
webpage

#to get  webpage content as html document using BeautifulSoup  method
soup  = BeautifulSoup(webpage.content,"html.parser")

#fetch product links
links = soup.find_all("a",attrs= {'class':'WKTcLC'})

#creates an empty list to store links
links_list = []

#loop for  extracting links from achor tag
for link in links:
    links_list.append(link.get('href'))
    d = {'title':[],'price':[],'ratings':[]}

#loop for  extracting product details from each links
for link in links_list:
    new_webpage = requests.get("https://www.flipkart.com"+link,headers=HEADERS)
    new_soup = BeautifulSoup(new_webpage.content,"html.parser")

    d['title'].append(get_title(new_soup))
    d['price'].append(get_price(new_soup))
    d['ratings'].append(get_ratings(new_soup))


amazon_df = pd.DataFrame(d)
#amazon_df['title'] = amazon_df['title'].replace(' ',np.nan,inplace=True)
#amazon_df = amazon_df.dropna(subset=['title'])
amazon_df.to_csv('amazonData.csv',index=False,header=True)

```
